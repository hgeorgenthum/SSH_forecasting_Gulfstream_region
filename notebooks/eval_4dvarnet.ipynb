```python
import sys
from pandas import Series, DataFrame
from tqdm import tqdm
```

```python
sys.path.append('..')
from src.eval_ose import eval_ose
```

```python
path_alongtrack = '../data/dt_gulfstream_c2_phy_l3_20161201-20180131_285-315_23-53.nc'
```

```python
time_min = '2017-02-01'
time_max = '2017-11-30'
```

```python
var_name = 'out'
```

```python
metrics = []
for leadtime in tqdm(range(7)):
    path_ds_4dvarnet = f'../data/ds_4dvarnet_leadtime_{leadtime}.nc'
    rmse, psd = eval_ose(path_alongtrack,
                         path_ds_4dvarnet,
                         var_name=var_name,
                         time_min=time_min,
                         time_max=time_max)
    metrics.append(Series({
        "(RMSE)": rmse,
        "x (km)": psd,
    }))
```

```python
print(DataFrame(metrics, range(7)).T.to_markdown())
```
